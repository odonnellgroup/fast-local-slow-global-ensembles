```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import math
```

```python
projdir = # NAME OF FOLDER THAT CONTAINS PROJECT
```

```python
# create objects to store spike counts and firing rates
deltat_list = ['0p1','1p0']
x = []
fr = []
for i in range(2):
    deltat = deltat_list[i] # string for timebin size
    x.append(np.load(projdir+'/npy/spike_count_frames/Krebs_'+deltat+'_spike_counts.npy',allow_pickle=True))
    fr.append(np.load(projdir+'/npy/firing_rate_frames/Krebs_'+deltat+'_firing.npy',allow_pickle=True))
```

```python
# extract all cell spike counts
cell_ids = []
t_ids = []
spikecountmat = []
tvec = []
for i in range(2):
    cell_ids.append(pd.unique(x[i].cell_id))
    ncells = len(cell_ids[i])
    t_ids.append(pd.unique(x[i].bin_start_time))
    nt = len(t_ids[i])
    spikecountmat.append(np.zeros((nt,ncells)))
    for j in range(ncells):
        spikecountmat[i][:,j] = x[i][x[i].cell_id==cell_ids[i][j]].spike_count
    tlength = np.max(x[i].bin_stop_time) - np.min(x[i].bin_stop_time)
    tvec.append(np.arange(0,tlength+0.5*x[i].bin_width[0],x[i].bin_width[0]))
```

```python
tstart = 1100
tend = 1150
nplot = 20
plt.plot(tvec[1][tstart:tend],spikecountmat[1][tstart:tend,0:nplot])
```

```python
# load correlations from Thomas's file
corrmat = []
for dt in range(2):
    filename = projdir+'/csv/analysis_frames/Krebs_'+deltat_list[dt]+'_analysis.csv'
    corrtable = np.loadtxt(open(filename, "rb"), delimiter=",", skiprows=1)
    lencorrtable = len(corrtable)

    # make correlation matrix
    corrmat.append(np.zeros((ncells,ncells)))
    for i in range(lencorrtable):
        celli_ind = np.where(cell_ids[dt]==corrtable[i,2])[0]
        cellj_ind = np.where(cell_ids[dt]==corrtable[i,5])[0]
        corrmat[dt][celli_ind,cellj_ind] = corrtable[i,0]
    for i in range(ncells):
        for j in range(ncells):
            if i>j:
                corrmat[dt][i,j] = corrmat[dt][j,i]
plt.imshow(corrmat[0],vmin=-0.05,vmax=0.25)
```

```python
# create region names list
cell_info = pd.read_csv(projdir+ '/csv/cell_info.csv')
uniqueregions = cell_info.cell_region.unique()
nregions = len(uniqueregions)
ncellsperregion = ncells/nregions
regions_ids = []
for i in range(nregions):
    for j in range(int(ncellsperregion)):
        regions_ids.append(uniqueregions[i])
```

```python
# create region colour dictionary
colour_list = ['red','green','yellow','magenta','blue','cyan','black','silver','turquoise','darkviolet','lime','goldenrod','cornflowerblue','gold','silver','darkred','violet','hotpink','peru','bisque']
region_colour_dict = {}
for i in range(len(uniqueregions)):
    region_colour_dict[uniqueregions[i]] = colour_list[i]
```

```python
# Scatter plot correlations from same region at small vs large timescales
limnum = 0.5
plt.plot([-limnum,limnum],[-limnum,limnum])
plt.plot([-limnum,limnum],[0.,0.],c='black')
plt.plot([0,0],[-limnum,limnum],c='black')
for i in range(ncells):
    sameinds = [ i for i in range(ncells) if regions_ids[i] == regions_ids[j] ]
    plt.scatter(corrmat[0][i,sameinds],corrmat[1][i,sameinds],s=2)
plt.xlabel("Correlation at fast timescale")
plt.ylabel("Correlation at slow timescale")
plt.xlim([-limnum,limnum])
plt.ylim([-limnum,limnum])
```

```python
# Scatter plot correlations from different region at small vs large timescales
plt.plot([-limnum,limnum],[-limnum,limnum])
plt.plot([-limnum,limnum],[0.,0.],c='black')
plt.plot([0,0],[-limnum,limnum],c='black')
for i in range(ncells):
    sameinds = [ i for i in range(ncells) if regions_ids[i] == regions_ids[j] ]
    diffinds = list(set(range(0,ncells)) - set(sameinds))
    plt.scatter(corrmat[0][i,diffinds],corrmat[1][i,diffinds],s=2)
plt.xlabel("Correlation at fast timescale")
plt.ylabel("Correlation at slow timescale")
plt.xlim([-limnum,limnum])
plt.ylim([-limnum,limnum])
```

```python
# Scatter plot correlations from same and different region at small vs large timescales, savefigs
limneg = -0.2
limpos = 0.5
maxlim=np.maximum(abs(limneg),limpos)
for j in range(len(uniqueregions)):
    for k in range(len(uniqueregions)):
        if j>k:
            plt.figure()
            plt.plot([-maxlim,maxlim],[-maxlim,maxlim],c='black',linewidth=0.5)
            plt.plot([-maxlim,maxlim],[0.,0.],c='black',linewidth=0.5)
            plt.plot([0,0],[-maxlim,maxlim],c='black',linewidth=0.5)
            sameinds = [ i for i in range(ncells) if regions_ids[i] == uniqueregions[j] ]
            otherinds = [ i for i in range(ncells) if regions_ids[i] == uniqueregions[k] ]
            plt.scatter(corrmat[0][sameinds[0]:sameinds[-1],otherinds[0]:otherinds[-1]],corrmat[1][sameinds[0]:sameinds[-1],otherinds[0]:otherinds[-1]],s=2,c='red')
            plt.scatter(corrmat[0][sameinds[0]:sameinds[-1],sameinds[0]:sameinds[-1]],corrmat[1][sameinds[0]:sameinds[-1],sameinds[0]:sameinds[-1]],s=2,c='black')
            plt.xlim([limneg,limpos])
            plt.ylim([limneg,limpos])
#             plt.show()
            plt.savefig('figs/'+uniqueregions[j]+'_'+uniqueregions[k]+'_corrscatter.png')
# plt.show()
```

```python
# find good example spike trains
epsilon_intra = 0.03
epsilon_inter = 0.03
ctarget_inter_long = 0.4
ctarget_intra_long = 0.4
ctarget_inter_short = 0.05
ctarget_intra_short = 0.4
goodintrapairs = [[0,0]]
goodinterpairs = [[0,0]]
for k in range(ncells):
    sameinds = [ i for i in range(ncells) if regions_ids[i] == regions_ids[j] ]
    for j in range(ncells):
        if j in sameinds:
            if np.abs(corrmat[0][k,j]-ctarget_intra_short)<epsilon_intra:
                if np.abs(corrmat[1][k,j]-ctarget_intra_long)<epsilon_intra:
                    goodintrapairs = np.append(goodintrapairs,[[k,j]],axis=0)
        else:
            if np.abs(corrmat[0][k,j]-ctarget_inter_short)<epsilon_inter:
                if np.abs(corrmat[1][k,j]-ctarget_inter_long)<epsilon_inter:
                    goodinterpairs = np.append(goodinterpairs,[[k,j]],axis=0)
```
